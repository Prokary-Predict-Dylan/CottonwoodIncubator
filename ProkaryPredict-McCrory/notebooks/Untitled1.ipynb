```python
import random

# Simple codon â†’ block mapping (illustrative)
BLOCKS = {
    'A': 'â–ˆ',  # Adenine
    'T': 'â–’',  # Thymine
    'C': 'â–“',  # Cytosine
    'G': 'â–‘'   # Guanine
}

def generate_random_dna(length=60):
    """Generate a random DNA sequence of A/T/C/G."""
    return ''.join(random.choice('ATCG') for _ in range(length))

def dna_to_barcode(seq):
    """Convert a DNA sequence into a block-coded visualization."""
    barcode = ''.join(BLOCKS.get(base, '?') for base in seq)
    return barcode

def main():
    print("ðŸ§¬  ProkaryPredict: DNA Barcode Visualizer")
    print("------------------------------------------\n")
    
    seq = generate_random_dna(90)
    barcode = dna_to_barcode(seq)
    
    print(f"Random DNA Sequence:\n{seq}\n")
    print("Visual Barcode Representation:\n")
    
    # Print barcode in lines of 45 blocks
    for i in range(0, len(barcode), 45):
        print(barcode[i:i+45])
    
    print("\nLegend:")
    for base, block in BLOCKS.items():
        print(f"  {base} â†’ {block}")

if __name__ == "__main__":
    main()
```
